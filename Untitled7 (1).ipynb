```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt
```

```python
# Inputs
distance = ctrl.Antecedent(np.arange(0, 51, 1), 'Distance')  # meters from no-parking sign
time = ctrl.Antecedent(np.arange(0, 25, 1), 'Time')          # hour of the day (0-24)

# Output
parking = ctrl.Consequent(np.arange(0, 101, 1), 'Parking')
```

```python
# Distance from No-Parking Zone
distance['very_close'] = fuzz.trimf(distance.universe, [0, 0, 10])
distance['close'] = fuzz.trimf(distance.universe, [5, 15, 25])
distance['far'] = fuzz.trimf(distance.universe, [20, 35, 50])

# Time of Parking
time['night'] = fuzz.trimf(time.universe, [0, 2, 6])
time['morning'] = fuzz.trimf(time.universe, [6, 9, 12])
time['afternoon'] = fuzz.trimf(time.universe, [12, 15, 18])
time['late'] = fuzz.trimf(time.universe, [18, 21, 24])

# Parking Legality
parking['illegal'] = fuzz.trimf(parking.universe, [60, 80, 100])
parking['borderline'] = fuzz.trimf(parking.universe, [30, 50, 70])
parking['legal'] = fuzz.trimf(parking.universe, [0, 20, 40])
```

```python
rule1 = ctrl.Rule(distance['very_close'] & time['morning'], parking['illegal'])
rule2 = ctrl.Rule(distance['very_close'] & time['night'], parking['borderline'])
rule3 = ctrl.Rule(distance['close'] & time['afternoon'], parking['borderline'])
rule4 = ctrl.Rule(distance['far'] & time['late'], parking['legal'])
rule5 = ctrl.Rule(distance['far'] & time['night'], parking['legal'])
rule6 = ctrl.Rule(distance['close'] & time['night'], parking['legal'])
rule7 = ctrl.Rule(distance['very_close'] & time['late'], parking['illegal'])
```

```python
parking_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7])
parking_sim = ctrl.ControlSystemSimulation(parking_ctrl)
```

```python
parking_sim.input['Distance'] = 8    # meters
parking_sim.input['Time'] = 10       # 10 AM

parking_sim.compute()
print(f"Parking Legality Score: {parking_sim.output['Parking']:.2f}")
parking.view(sim=parking_sim)
```

```python
distance.view()
time.view()
parking.view()
```

```python
# Meshgrid for inputs
x_range = np.arange(0, 51, 5)
y_range = np.arange(0, 25, 1)
x, y = np.meshgrid(x_range, y_range)
z = np.zeros_like(x, dtype=float)

# Simulate parking decision surface
for i in range(len(x_range)):
    for j in range(len(y_range)):
        sim = ctrl.ControlSystemSimulation(parking_ctrl)
        sim.input['Distance'] = x[j, i]
        sim.input['Time'] = y[j, i]
        sim.compute()
    try:
       sim.compute()
       z[j, i] = sim.output['Parking']
    except:
       z[j, i] = np.nan  # Handles cases where fuzzy logic fails


# 3D Plot
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='coolwarm')
ax.set_xlabel('Distance from No-Parking (m)')
ax.set_ylabel('Time of Day (hr)')
ax.set_zlabel('Parking Legality')
ax.set_title('Fuzzy Parking Classifier Surface')
plt.show()
```
